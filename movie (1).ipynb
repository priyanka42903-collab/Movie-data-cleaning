# Movie Dataset Cleaning and Preprocessing

This project focuses on cleaning and preparing a movie dataset for further analysis.
The notebook demonstrates handling missing values, correcting data types, and basic feature engineering using Python and Pandas.

```python
df = pd.read_csv('Tmdb top rated movies.csv')
```

```python
df
```

### To make the movie ratings easier to interpret, I am categorizing the `vote_average` scores into three groups:
- Poor: rating below 6  
- Average: rating between 6 and 7.5  
- Good: rating above 7.5  

A new column will be created called `rating_category` based on these rules.

```python
def rating_category(v):
    if v < 6:
        return 'Poor'
    elif v < 7.5:
        return 'Average'
    else:
        return 'Good'

df['rating_category'] = df['vote_average'].apply(rating_category)
```

```python
df
```

### Removing unnecessary columns such as unnamed,original_language,original_title,vote_average

```python
data= df[['id','title','overview','popularity','release_date','vote_count','rating_category']]
```

```python
data
```

```python
df.describe()
```

```python
data.info()
```

### Checking repeated titles

```python
data['title'].value_counts()
```

### Sorting the data from the newest to the oldest release_date

```python
data = data.sort_values('release_date',ascending=False)
```

```python
data
```

### Removing rows with same title and id and keeping only the first occurrence

```python
data = data.drop_duplicates(subset=['title','id'], keep='first')
```

```python
data
```

```python
data.info()
```

### Checking null values in each column 

```python
data.isnull().sum()
```

### Filling null values

```python
data= data.copy()
data['overview'] = data['overview'].fillna("Not Available")
```

### Converting release_date to datetime format

```python
data['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
data = data.dropna(subset=['release_date'])
```

### Rechecking null values

```python
data.isnull().sum()
```

### Converting float data type into integer data type

```python
data['popularity'] = data['popularity'].round().astype(int)
```

```python
data.info()
```
